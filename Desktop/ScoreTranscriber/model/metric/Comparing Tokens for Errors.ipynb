```python
def midi_note_to_name(notenum):
    # convert midi pitch num into xml note num eg 69 --> A4 
    notes = ["A", "Bb", "B", "C", "C#", "D", "Eb", "E", "F", "F#", "G", "G#"]
    octave = int (notenum / 12) - 1
    notenum -= 21
    note = notes[notenum % 12]

    return note+str(octave) 
```
